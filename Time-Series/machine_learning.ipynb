```python
import pandas as pd
import warnings
from cesium import datasets
import matplotlib.pyplot as plt
import numpy as np
from cesium.featurize import featurize_time_series as ft

warnings.filterwarnings('ignore')
```

```python
eeg = datasets.fetch_andrzejak()
```

```python
eeg['measurements'][0].shape, eeg['classes'].shape, np.unique(eeg['classes'])
```

```python
j = 1
cmap = plt.cm.get_cmap('Set1')
for i in [0,300,450]:
    plt.cm.get_cmap()
    plt.subplot(3,1,j)
    plt.plot(eeg['measurements'][i], c=cmap(j))
    plt.legend(eeg['classes'][i])
    j += 1
```

```python
features_to_use = ['amplitude','percent_beyond_1_std','percent_close_to_median','skew','max_slope']

fset_cesium = ft(times=eeg['times'], values=eeg['measurements'], errors=None, features_to_use=features_to_use, scheduler=None)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
```

## Bootstraping

1. We estimate $f(x)$ from population using our size-$n$ sample PDF $\hat{f}(x)$
2. We take $M$ random samples with replacement from $\hat{f}(x)$ (with size $n$), and generate an approximation of $\hat{f}(\hat{\theta})$ to get estimations of $\hat{\theta}$

Obs.: Sampling scheme from the population and to estimate $\hat{f}(\hat{\theta})$ should be the same

### Non-parametric Bootstraping

Bootstraping the **mean** of a $n$-size sample taken from a normal distribution with $10^4$ boots:

```python
n = 50
mu = 32
std = 5
sample_data = np.random.normal(loc=mu, scale=std, size=n)
M = 10**4
boots = np.zeros(M)

for i in range(M):
    resample = np.random.choice(a=sample_data, size=n, replace=True)
    boots[i] = np.mean(resample)

sn.histplot(boots)
plt.xlabel('Mean values')
plt.show()
```

Bootstraping the **variance** of a $n$-size sample taken from a normal distribution with $10^4$ boots:

```python
n = 50
mu = 32
std = 5
sample_data = np.random.normal(loc=mu, scale=std, size=n)
M = 10**4
boots = np.zeros(M)

for i in range(M):
    resample = np.random.choice(a=sample_data, size=n, replace=True)
    boots[i] = np.var(resample)

sn.histplot(boots)
plt.xlabel('Variance values')
plt.show()
```

### Parametric Bootstraping

```python
n = 50
K = 10**4
mu = 32
std = 5
sample_data = np.random.normal(loc=mu, scale=std, size=n)
boots = np.zeros(K)
xbar = np.mean(sample_data)
s = np.std(sample_data)

for k in range(K):
    resample = np.random.normal(loc=xbar, scale=s, size=n)
    boots[k] = np.mean(resample)

boots_df = pd.DataFrame(boots)
display(boots_df.describe())

sn.histplot(boots, color='m')
plt.xlabel('Mean values')
plt.show()
```
